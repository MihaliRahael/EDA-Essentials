```python
# Next we will process Arrival time. Entries are not uniform. We can remove date portion since we already have date info.

final_df['Arrival_Time'].str.split(' ')
```

```python
# we need to take only time which is the first index of result. We can do same operation using lambda as below.
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
# we will use lambda function for making changes

final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
# Now we will split hour and min
final_df['Arrival_hour']= final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_min']= final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(5)
```

```python
# now we will process dep time

final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
# Our intention is to predict the price. We need to think whether the route feature is so important in predicting price.
# Because route feature gives info about stops in between source and dest. Total stops gives the same info of # of stops.
# So we will process total stops and avoid route
# Single Nan value is there.After seeing the route is Delhi-->Cochin, we assume it will take 1 stop in between. So assign nan=1

final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':1, '4 stops':4})
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```
